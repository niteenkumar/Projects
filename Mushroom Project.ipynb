# Model Building

```python
y=data['class']
x=data.drop('class',axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8)
```

as we have a large data set so we can feed 80% of data for training the model

Let's some models

# Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(random_state=0,max_depth=5)
dt.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
print(dt.score(x_train,y_train))
print(accuracy_score(y_test,dt.predict(x_test)))
```

so for Decision Tree model,
testing score is 0.979227573472842

prediction accuracy score is 0.9796923076923076

# Logistic Regression model

```python
from sklearn.linear_model import LogisticRegression
log_model=LogisticRegression()
log_model.fit(x_train,y_train)
```

```python
print(log_model.score(x_train,y_train))
print(accuracy_score(y_test,log_model.predict(x_test)))
```

so for Logistic Regression model, testing score is 0.9550700107708878

prediction accuracy score is 0.9612307692307692

So from the above two models' score we can clearly see that Decision tree model is giving good score.
